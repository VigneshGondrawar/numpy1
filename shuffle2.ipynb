```python
#Syntax:	numpy.random.shuffle(ndarrayobject)
import numpy as np
a=([10,20,30,40,50])
print("before shuffle() content of a={} type={} and ID={}".format(a,type(a),id(a)))
np.random.shuffle(a)
print("After Shuffle() content of a={} type={} and ID={}".format(a,type(a),id(a)))
```

```python
for i in range(len(a)):
  np.random.shuffle(a)
  print(a)
```

```python
for i in range(len(a)):
  np.random.shuffle(a)
  print(a,id(a))
```

```python
#Syntax:varname=numpy.random.permutation(ndarrayobject)
a=np.array([10,20,30,40,50])
print("before permutation content of a={} type={} and ID={}".format(a,type(a),id(a)))
b=np.random.permutation(a)
print("After permutation content of a={} type={} and ID={}".format(a,type(a),id(a)))
print("After permutation content of a={} type={} and ID={}".format(b,type(b),id(b)))
```

```python
b=np.random.permutation(a)
print("After permutation content of a={} type={} and ID={}".format(b,type(b),id(b)))
```

```python
b=np.random.permutation(a)
print("After permutation content of a={} type={} and ID={}".format(b,type(b),id(b)))
```

```python
for i in range(len(a)):
  b=np.random.permutation(a)
  print(b,id(b))
```
