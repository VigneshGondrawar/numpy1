```python
import numpy as np
a=np.array(range(10,31,2))
print(a,type(a))
print("Dimension of a=",a.ndim)
```

```python
for i in a:
  print(i)
```

```python
#Iterate the 2-D Array data
a=np.array(range(10,33,2)).reshape(3,4)
print(a,type(a))
print("Dimension of a=",a.ndim)
```

```python
for i in a:
  for j in i:
    print(j)
```

```python
#Iterate the 3-D Array data
a=np.arange(2*3*3).reshape(2,3,3)
print(a,type(a))
print("Dimension of a=",a.ndim)
```

```python
#Iterate the 3-D Array data
for i in a:
  for j in i:
    for g in j:
      print(g)
```
